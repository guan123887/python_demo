- (9)以每个用户7月和8月用电数据为同一批统计样本，3月和4月用电数据为另一批统计样本，分别计算这两批样本之间的总体和(sum)之比、均值(mean)之比、最大值(max)之比和最小值(min)之比.
- (10)将上述统计的所有特征合并在一张表中显示出来.

```python
import pandas as pd
import numpy as np
```

```python
# (1)将数据进行转置，转置后型如eg.csv，缺失值用NAN代替.
## parse_dates为解析日期，index_col为列表表示按照0列进行索引，不满足按照1列进行索引
df = pd.read_csv('./data/data.csv', parse_dates=True, index_col=[0, 1])
df
```

```python
df = df.unstack().KWH # 行列转换
df
```

```python
df.fillna(np.nan, inplace=True)
df
```

```python
# (2)对数据中的异常值进行识别并用NA代替。
## 识别异常值
for col in range(df.shape[1]):
  # 当前值与均值的差值绝对值大于三倍标准差视为异常值
  df.iloc[:, col][(df.iloc[:, col]-df.mean(axis=1)).abs()>df.std(axis=1)*3] = np.nan
df
```

```python
# (3)计算每个用户用电数据的基本统计量，包括最大值、最小值、均值、中位数、和、方差、偏度、峰值。(不包括空值)
df.agg(['max', 'min', 'mean', 'median', 'sum', 'var', 'skew', 'kurt'], axis=1)
```

```python
# (4)每个用户用电数据按日差分，并计算差分结果的基本统计量，统计量同上述第3问.

## diff方法为计算差分
df_diff = df.diff(axis=1)
df_diff_max = df_diff.max(axis=1) # 最大值
df_diff_min = df_diff.min(axis=1) # 最小值
df_diff_mean = df_diff.mean(axis=1) # 均值
df_diff_median = df_diff.median(axis=1) # 中位数
df_diff_sum = df_diff.sum(axis=1) # 和
df_diff_var = df_diff.var(axis=1) # 方差
df_diff_skew = df_diff.skew(axis=1) # 偏度
df_diff_kurt = df_diff.kurt(axis=1) # 峰度

## 或者
df.diff(axis=1).agg(['max', 'min', 'mean', 'median', 'sum', 'var', 'skew', 'kurt'], axis=1)
```

```python
# (5)计算每个用户用电数据的5%分位数.
df_quantile = df.quantile([.05], axis=1)
df_quantile
```

```python
# (6)对每个用户的用电数据按周求和并差分(一周7天)，计算差分结果的基本统计量，统计量同第3问.
df_week_diff = df.groupby(df.columns.week, axis=1).sum().diff(axis=1)
df_week_diff.agg(['max', 'min', 'mean', 'median', 'sum', 'var', 'skew', 'kurt'], axis=1)
```

```python
# (7)每个用户在一段时间内会有用电数据最大值，统计用电数据大于“0.9×最大值”的天数
# df_per10 = pd.Series([(df.iloc[row,:]>(df.max(axis=1)*0.9).iloc[row]).sum() for row in range(df.shape[0])], index=range(1, 201))
# df_per10

# 或者
df_per10 = df.max(axis=1) * 0.9
df[df.sub(df_per10, axis=0)>0].count(axis=1)
```
