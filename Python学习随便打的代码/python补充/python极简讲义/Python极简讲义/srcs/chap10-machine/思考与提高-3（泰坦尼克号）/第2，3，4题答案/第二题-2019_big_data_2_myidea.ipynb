```python
#导入用户用电量数据
import pandas as pd

data=pd.read_csv("./用户用电量数据/data.csv")

data.info()
```

# （1）将数据进行转置，转置为型如eg.csv为所示（缺失的用NAN代替）

```python
import numpy as np

k1 = np.array(data[data.CONS_NO==1][["KWH"]])
k2 = np.array(data[data.CONS_NO==1][["DATA_DATE"]])
g = pd.Series(k1.reshape(len(k1),),index=k2.reshape(len(k2),))
df = g.to_frame().T

for i in range(2,201):
    k3 =np.array(data[data.CONS_NO==i][["KWH"]])
    k4 =np.array(data[data.CONS_NO==i][["DATA_DATE"]])
    g_ = pd.Series(k3.reshape(len(k3),),index=k4.reshape(len(k4),))
    df= df.append(g_,ignore_index=True)
a=np.arange(1,201)
df.insert(0,'CONS_NO',a)
# df.T.info()
df.fillna("NAN")
```

# 2)对数据中的异常数据进行识别并用NA代替（5分）

```python
#箱线图处理异常数据
import numpy as np
d2 = df.iloc[:,1:].T
for i in d2.columns:
    iqr = d2.iloc[:,i].quantile(0.75) - d2.iloc[:,i].quantile(0.25)
    q_l = d2.iloc[:,i].quantile(0.25) - 1.5 * iqr
    q_u = d2.iloc[:,i].quantile(0.75) +1.5 * iqr
    for j in range(len(d2.index)):
        if (d2.iloc[j,i]<q_l) | (d2.iloc[j,i] > q_u):
                d2.iloc[j,i] = np.nan
d2
```

# 3 求每个用电用户的基本统计量:

```python
statistic=pd.DataFrame(columns=["sum","mean","max","std","var","median","skew","kurt"])
# d2.mean(axis=0)
statistic["sum"]=d2.sum()
statistic["mean"]=d2.mean()
statistic["max"]=d2.max()
statistic["std"]=d2.std()
statistic["var"]=d2.var()
statistic["median"]=d2.median()
statistic["skew"]=d2.skew()
statistic["kurt"]=d2.kurt()
statistic.T
```

# 6)每个用户按周求和并差分（一周7天），并求取差分结果的基本统计量，统计量同3。

```python
data.head()
```

```python
date_new=data.fillna(np.nan)#填充缺失值
#按周统计
column=np.arange(1,201)
date_d=date_new[["DATA_DATE","KWH"]][data["CONS_NO"]==1]
date_d.columns = ["date","kwh"]
date_d['date'] = pd.to_datetime(date_d['date'])
date_d = date_d.set_index("date")
date_temp=date_d.resample('w').sum().diff()
column=np.arange(1,201)
index1 = date_temp.index
date_df = pd.DataFrame(index=index1,columns=column)
date_df[1] = date_temp
#增加其他用户数据
for i in range(2,201):
    date_d=date_new[["DATA_DATE","KWH"]][data["CONS_NO"]==i]
    date_d.columns = ["date","kwh"]
    date_d['date'] = pd.to_datetime(date_d['date'])
    date_d = date_d.set_index("date")
    date_temp=date_d.resample('w').sum().diff()
    date_df[i] = date_temp
```

```python
date_df
```

```python
#按周求和差分后求统计量
diff_week_statistic=pd.DataFrame(columns=["sum","mean","max","std","var","median","skew","kurt"])
# d2.mean(axis=0)
diff_week_statistic["sum"]=date_df.sum()
diff_week_statistic["mean"]=date_df.mean()
diff_week_statistic["max"]=date_df.max()
diff_week_statistic["std"]=date_df.std()
diff_week_statistic["var"]=date_df.var()
diff_week_statistic["median"]=date_df.median()
diff_week_statistic["skew"]=date_df.skew()
diff_week_statistic["kurt"]=date_df.kurt()
diff_week_statistic.T
```
